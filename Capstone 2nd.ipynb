```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import widgets, Layout, HBox, VBox
from IPython.display import display, clear_output
from io import BytesIO
import matplotlib as mpl

mpl.rcParams['font.sans-serif'] = ['Arial']
mpl.rcParams['axes.unicode_minus'] = False

df = None
fig = None

def upload_file(change):
    global df
    if change['new']:
        uploaded_file = change['new'][0]
        content = uploaded_file['content']
        try:
            df = pd.read_csv(BytesIO(content), encoding='utf-8')
        except UnicodeDecodeError:
            try:
                df = pd.read_csv(BytesIO(content), encoding='latin1')
            except UnicodeDecodeError:
                df = pd.read_csv(BytesIO(content), encoding='ISO-8859-1')
        file_label.value = f"File uploaded: {uploaded_file['name']}"

def generate_visualization(change):
    global fig
    if df is not None:
        output.clear_output()
        with output:
            if plot_type.value == 'Bar Plot':
                required_columns = ['Soils_C', 'Soils_A', 'Soils_error', 'Soils_DISP', 'Soils_E',
                                    'Sulphate_C', 'Sulphate_A', 'Sulphate_error', 'Sulphate_DISP', 'Sulphate_E',
                                    'Diesel vehicles_C', 'Diesel vehicles_A', 'Diesel vehicles_error', 'Diesel vehicles_DISP', 'Diesel vehicles_E',
                                    'Marine aerosol_C', 'Marine aerosol_A', 'Marine aerosol_error', 'Marine aerosol_DISP', 'Marine aerosol_E',
                                    'Biomass burning_C', 'Biomass burning_A', 'Biomass burning_error', 'Biomass burning_DISP', 'Biomass burning_E',
                                    'Petrol vehicles_C', 'Petrol vehicles_A', 'Petrol vehicles_error', 'Petrol vehicles_DISP', 'Petrol vehicles_E',
                                    'Construction_C', 'Construction_A', 'Construction_error', 'Construction_DISP', 'Construction_E']

                missing_columns = [col for col in required_columns if col not in df.columns]
                if missing_columns:
                    print(f"The uploaded CSV file does not contain the following required columns: {', '.join(missing_columns)}")
                    return

                fig, ax = plt.subplots(7, 1, figsize=(9, 14), sharex=True)
                fig.subplots_adjust(hspace=0.107, left=0.333, bottom=0.08, top=0.945)

                for i in range(7):
                    ax[i].set_yscale('log')

                ax00 = ax[0].twinx()
                ax01 = ax[1].twinx()
                ax02 = ax[2].twinx()
                ax03 = ax[3].twinx()
                ax04 = ax[4].twinx()
                ax05 = ax[5].twinx()
                ax06 = ax[6].twinx()

                for a in ax:
                    a.set_ylim(0.0001, 2)
                for a in [ax00, ax01, ax02, ax03, ax04, ax05, ax06]:
                    a.set_ylim(0, 100)
                    a.set_yticks([0, 20, 40, 60, 80])

                ls = 'dotted'
                for a in ax:
                    a.grid(axis="x", linestyle=ls, linewidth=1, zorder=2, color="gray")

                species = df['Species'].tolist()

                Soils_C = df['Soils_C'].tolist()
                Soils_A = df['Soils_A'].tolist()
                Soils_error = df['Soils_error'].tolist()
                Soils_DISP = df['Soils_DISP'].tolist()

                Sulphate_C = df['Sulphate_C'].tolist()
                Sulphate_A = df['Sulphate_A'].tolist()
                Sulphate_error = df['Sulphate_error'].tolist()
                Sulphate_DISP = df['Sulphate_DISP'].tolist()

                Diesel_vehicles_C = df['Diesel vehicles_C'].tolist()
                Diesel_vehicles_A = df['Diesel vehicles_A'].tolist()
                Diesel_vehicles_error = df['Diesel vehicles_error'].tolist()
                Diesel_vehicles_DISP = df['Diesel vehicles_DISP'].tolist()

                Marine_aerosol_C = df['Marine aerosol_C'].tolist()
                Marine_aerosol_A = df['Marine aerosol_A'].tolist()
                Marine_aerosol_error = df['Marine aerosol_error'].tolist()
                Marine_aerosol_DISP = df['Marine aerosol_DISP'].tolist()

                Biomass_burning_C = df['Biomass burning_C'].tolist()
                Biomass_burning_A = df['Biomass burning_A'].tolist()
                Biomass_burning_error = df['Biomass burning_error'].tolist()
                Biomass_burning_DISP = df['Biomass burning_DISP'].tolist()

                Petrol_vehicles_C = df['Petrol vehicles_C'].tolist()
                Petrol_vehicles_A = df['Petrol vehicles_A'].tolist()
                Petrol_vehicles_error = df['Petrol vehicles_error'].tolist()
                Petrol_vehicles_DISP = df['Petrol vehicles_DISP'].tolist()

                Construction_C = df['Construction_C'].tolist()
                Construction_A = df['Construction_A'].tolist()
                Construction_error = df['Construction_error'].tolist()
                Construction_DISP = df['Construction_DISP'].tolist()

                Soils_E = df['Soils_E'].tolist()
                Sulphate_E = df['Sulphate_E'].tolist()
                Diesel_vehicles_E = df['Diesel vehicles_E'].tolist()
                Marine_aerosol_E = df['Marine aerosol_E'].tolist()
                Biomass_burning_E = df['Biomass burning_E'].tolist()
                Petrol_vehicles_E = df['Petrol vehicles_E'].tolist()
                Construction_E = df['Construction_E'].tolist()

                Soils_A = np.array(Soils_A)
                Soils_error = np.array(Soils_error)
                Soils_DISP = np.array(Soils_DISP)

                Sulphate_A = np.array(Sulphate_A)
                Sulphate_error = np.array(Sulphate_error)
                Sulphate_DISP = np.array(Sulphate_DISP)

                Diesel_vehicles_A = np.array(Diesel_vehicles_A)
                Diesel_vehicles_error = np.array(Diesel_vehicles_error)
                Diesel_vehicles_DISP = np.array(Diesel_vehicles_DISP)

                Marine_aerosol_A = np.array(Marine_aerosol_A)
                Marine_aerosol_error = np.array(Marine_aerosol_error)
                Marine_aerosol_DISP = np.array(Marine_aerosol_DISP)

                Biomass_burning_A = np.array(Biomass_burning_A)
                Biomass_burning_error = np.array(Biomass_burning_error)
                Biomass_burning_DISP = np.array(Biomass_burning_DISP)

                Petrol_vehicles_A = np.array(Petrol_vehicles_A)
                Petrol_vehicles_error = np.array(Petrol_vehicles_error)
                Petrol_vehicles_DISP = np.array(Petrol_vehicles_DISP)

                Construction_A = np.array(Construction_A)
                Construction_error = np.array(Construction_error)
                Construction_DISP = np.array(Construction_DISP)

                factors = [
                    (Soils_C, Soils_A, Soils_error, Soils_DISP, Soils_E, ax[0], ax00, 'Soil/Road dust'),
                    (Sulphate_C, Sulphate_A, Sulphate_error, Sulphate_DISP, Sulphate_E, ax[1], ax01, 'Sulphate/Marine diesel'),
                    (Diesel_vehicles_C, Diesel_vehicles_A, Diesel_vehicles_error, Diesel_vehicles_DISP, Diesel_vehicles_E, ax[2], ax02, 'Diesel vehicles'),
                    (Marine_aerosol_C, Marine_aerosol_A, Marine_aerosol_error, Marine_aerosol_DISP, Marine_aerosol_E, ax[3], ax03, 'Sea salt'),
                    (Biomass_burning_C, Biomass_burning_A, Biomass_burning_error, Biomass_burning_DISP, Biomass_burning_E, ax[4], ax04, 'Biomass burning'),
                    (Petrol_vehicles_C, Petrol_vehicles_A, Petrol_vehicles_error, Petrol_vehicles_DISP, Petrol_vehicles_E, ax[5], ax05, 'Petrol vehicles'),
                    (Construction_C, Construction_A, Construction_error, Construction_DISP, Construction_E, ax[6], ax06, 'Construction')
                ]

                for conc, asym, err, disp, expl, ax_, ax_twin, label in factors:
                    ax_.bar(species, conc, color='lightskyblue', edgecolor='black', linewidth=1, width=0.6)
                    ax_.scatter(species, asym, edgecolor="firebrick", alpha=1, s=80, c='white', linewidth=1.5, zorder=2)
                    ax_.errorbar(species, asym, yerr=[np.abs(asym - err), np.abs(disp - asym)], linestyle='', fmt='|', capsize=0.5, capthick=1, color="black", linewidth=1.5, solid_capstyle='butt')
                    ax_twin.scatter(species, expl, color="black", s=60)
                    ax_.text(10, 0.05, label, fontsize=15)

                for a in ax:
                    a.tick_params(labelsize=12)

                for a in [ax00, ax01, ax02, ax03, ax04, ax05, ax06]:
                    a.tick_params(labelsize=12)

                ax[4].set_ylabel('Concentration(ng/ng)', fontsize=16)
                ax04.set_ylabel('% Explained Variation', fontsize=16)

                ax[6].set_xticks(np.arange(len(species)))
                ax[6].set_xticklabels(species, fontsize=12)

                ax[0].set_xticks(np.arange(len(species)))
                ax[0].set_xticklabels(species, fontsize=12)
                ax[0].xaxis.set_ticks_position('top')

            elif plot_type.value == 'Pie Chart':
                fig, ax = plt.subplots()
                df.iloc[0].plot.pie(ax=ax, autopct='%1.1f%%', startangle=90)
                ax.axis('equal')
            
            elif plot_type.value == 'Stacked Bar Chart':
                fig, ax = plt.subplots(figsize=(10, 7))
                df.plot(kind='bar', stacked=True, ax=ax)
                ax.set_title('PM10 Source Contributions Over Years')
                ax.set_xlabel('Source')
                ax.set_ylabel('Percentage')
                ax.legend(title='Year')
            
            plt.show()
    else:
        with output:
            clear_output()
            print("Please upload a CSV file first.")

def save_as_jpeg(change):
    if fig is not None:
        buf = BytesIO()
        fig.savefig(buf, format='jpeg')
        buf.seek(0)
        with open("visualization.jpg", "wb") as f:
            f.write(buf.read())
        with output:
            clear_output()
            print("Image saved successfully!")
    else:
        with output:
            clear_output()
            print("Please generate a visualization first.")

upload_button = widgets.FileUpload(accept='.csv', multiple=False)
upload_button.observe(upload_file, names='value')

file_label = widgets.Label(value="Upload a CSV file")

generate_button = widgets.Button(description="Generate", button_style='info')
generate_button.on_click(generate_visualization)

save_button = widgets.Button(description="Save as JPEG Image", button_style='success')
save_button.on_click(save_as_jpeg)

plot_type = widgets.RadioButtons(
    options=['Bar Plot', 'Pie Chart', 'Stacked Bar Chart'],
    description='Plot Type:',
    disabled=False
)

output = widgets.Output()

controls = VBox([HBox([file_label, upload_button]),
                 HBox([generate_button, plot_type, save_button])])

display(controls, output)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from io import BytesIO
import base64
import numpy as np
import plotly.express as px

# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("PM10 Source Contributions Dashboard"),
    
    # File upload component
    dcc.Upload(
        id='upload-data',
        children=html.Div([
            'Drag and Drop or ',
            html.A('Select a CSV File')
        ]),
        style={
            'width': '100%',
            'height': '60px',
            'lineHeight': '60px',
            'borderWidth': '1px',
            'borderStyle': 'dashed',
            'borderRadius': '5px',
            'textAlign': 'center',
            'margin': '10px'
        },
        multiple=False
    ),
    html.Div(id='file-upload-info'),
    
    # Radio buttons for plot type selection
    dcc.RadioItems(
        id='plot-type',
        options=[
            {'label': 'Bar Plot', 'value': 'Bar Plot'},
            {'label': 'Pie Chart', 'value': 'Pie Chart'},
            {'label': 'Stacked Bar Chart', 'value': 'Stacked Bar Chart'}
        ],
        value='Bar Plot',
        labelStyle={'display': 'inline-block'}
    ),
    
    # Graph component to display the plot
    dcc.Graph(id='pm10-plot'),

    # Button to save the plot as a JPEG
    html.Button('Save as JPEG', id='save-button', n_clicks=0),
    dcc.Download(id='download-image')
])

def parse_contents(contents, filename):
    content_type, content_string = contents.split(',')
    decoded = BytesIO(base64.b64decode(content_string))
    try:
        if 'csv' in filename:
            # upload a CSV file
            df = pd.read_csv(decoded)
        else:
            return None
    except Exception as e:
        print(e)
        return None
    return df

def safe_get_list(df, column_name):
    if column_name in df.columns:
        return df[column_name].tolist()
    else:
        return [0] * len(df) 

@app.callback(
    Output('file-upload-info', 'children'),
    Output('pm10-plot', 'figure'),
    Input('upload-data', 'contents'),
    State('upload-data', 'filename'),
    Input('plot-type', 'value')
)
def update_output(contents, filename, plot_type):
    if contents is None:
        return '', {}

    df = parse_contents(contents, filename)

    if df is None:
        return 'There was an error processing this file.', {}

    if plot_type == 'Bar Plot':
        # Define the species list
        species = ["H", "BC", "Na", "Mg", "Al", "Si", "S", "Cl", "K", "Ca", "Ti", "V", "Mn", "Fe", "Co", "Ni", "Cu", "Zn", "As", "Ba"]

        # Extract data for each factor from the dataset
        Soils_C = safe_get_list(df, 'Soils_C')
        Soils_A = safe_get_list(df, 'Soils_A')
        Soils_error = safe_get_list(df, 'Soils_error')
        Soils_DISP = safe_get_list(df, 'Soils_DISP')

        Sulphate_C = safe_get_list(df, 'Sulphate_C')
        Sulphate_A = safe_get_list(df, 'Sulphate_A')
        Sulphate_error = safe_get_list(df, 'Sulphate_error')
        Sulphate_DISP = safe_get_list(df, 'Sulphate_DISP')

        Diesel_vehicles_C = safe_get_list(df, 'Diesel vehicles_C')
        Diesel_vehicles_A = safe_get_list(df, 'Diesel vehicles_A')
        Diesel_vehicles_error = safe_get_list(df, 'Diesel vehicles_error')
        Diesel_vehicles_DISP = safe_get_list(df, 'Diesel vehicles_DISP')

        Marine_aerosol_C = safe_get_list(df, 'Marine aerosol_C')
        Marine_aerosol_A = safe_get_list(df, 'Marine aerosol_A')
        Marine_aerosol_error = safe_get_list(df, 'Marine aerosol_error')
        Marine_aerosol_DISP = safe_get_list(df, 'Marine aerosol_DISP')

        Biomass_burning_C = safe_get_list(df, 'Biomass burning_C')
        Biomass_burning_A = safe_get_list(df, 'Biomass burning_A')
        Biomass_burning_error = safe_get_list(df, 'Biomass_burning_error')
        Biomass_burning_DISP = safe_get_list(df, 'Biomass_burning_DISP')

        Petrol_vehicles_C = safe_get_list(df, 'Petrol vehicles_C')
        Petrol_vehicles_A = safe_get_list(df, 'Petrol vehicles_A')
        Petrol_vehicles_error = safe_get_list(df, 'Petrol vehicles_error')
        Petrol_vehicles_DISP = safe_get_list(df, 'Petrol vehicles_DISP')

        Construction_C = safe_get_list(df, 'Construction_C')
        Construction_A = safe_get_list(df, 'Construction_A')
        Construction_error = safe_get_list(df, 'Construction_error')
        Construction_DISP = safe_get_list(df, 'Construction_DISP')

        # Extract explained variation data
        Soils_E = safe_get_list(df, 'Soils_E')
        Sulphate_E = safe_get_list(df, 'Sulphate_E')
        Diesel_vehicles_E = safe_get_list(df, 'Diesel vehicles_E')
        Marine_aerosol_E = safe_get_list(df, 'Marine aerosol_E')
        Biomass_burning_E = safe_get_list(df, 'Biomass_burning_E')
        Petrol_vehicles_E = safe_get_list(df, 'Petrol vehicles_E')
        Construction_E = safe_get_list(df, 'Construction_E')

        # Create the subplots
        fig = make_subplots(rows=7, cols=1, shared_xaxes=True, subplot_titles=("Soil/Road dust", "Sulphate/Marine diesel", "Diesel vehicles", 
                                                                               "Sea salt", "Biomass burning", "Petrol vehicles", "Construction"))
        factors = [
            (Soils_C, Soils_A, Soils_error, Soils_DISP, Soils_E, 1, 'Soil/Road dust'),
            (Sulphate_C, Sulphate_A, Sulphate_error, Sulphate_DISP, Sulphate_E, 2, 'Sulphate/Marine diesel'),
            (Diesel_vehicles_C, Diesel_vehicles_A, Diesel_vehicles_error, Diesel_vehicles_DISP, Diesel_vehicles_E, 3, 'Diesel vehicles'),
            (Marine_aerosol_C, Marine_aerosol_A, Marine_aerosol_error, Marine_aerosol_DISP, Marine_aerosol_E, 4, 'Sea salt'),
            (Biomass_burning_C, Biomass_burning_A, Biomass_burning_error, Biomass_burning_DISP, Biomass_burning_E, 5, 'Biomass burning'),
            (Petrol_vehicles_C, Petrol_vehicles_A, Petrol_vehicles_error, Petrol_vehicles_DISP, Petrol_vehicles_E, 6, 'Petrol vehicles'),
            (Construction_C, Construction_A, Construction_error, Construction_DISP, Construction_E, 7, 'Construction')
        ]

        for conc, asym, err, disp, expl, row, label in factors:
            fig.add_trace(go.Bar(x=species, y=conc, name=f'{label} C', marker_color='lightskyblue'), row=row, col=1)
            fig.add_trace(go.Scatter(x=species, y=asym, mode='markers', marker=dict(color='white', size=8, line=dict(color='firebrick', width=2)), name=f'{label} A'), row=row, col=1)
            fig.add_trace(go.Scatter(x=species, y=expl, mode='markers', marker=dict(color='black', size=6), name=f'{label} E'), row=row, col=1)
            fig.add_trace(go.Scatter(x=species, y=disp, mode='markers', marker=dict(color='black', size=6), name=f'{label} DISP'), row=row, col=1)

        fig.update_layout(height=1200, title_text="PM10 Source Contributions")
        fig.update_yaxes(type="log", range=[-4, 0.3], title_text="Concentration (ng/ng)")
        fig.update_xaxes(tickangle=45)

    elif plot_type == 'Pie Chart':
        sources = df.columns
        percentages = df.iloc[0]
        pie_data = pd.DataFrame({'Source': sources, 'Percentage': percentages})
        fig = px.pie(pie_data, values='Percentage', names='Source', title='PM10 Source Contributions',
                     labels={'Source': 'PM10 Sources', 'Percentage': 'Contribution (%)'})
    
    elif plot_type == 'Stacked Bar Chart':
        numeric_df = df.select_dtypes(include=[np.number])
        fig = px.bar(numeric_df, x=numeric_df.index, y=numeric_df.columns, title='PM10 Source Contributions Over Years', labels={'value': 'Percentage', 'index': 'Source'})
        fig.update_layout(barmode='stack')

    return f'File uploaded: {filename}', fig

@app.callback(
    Output('download-image', 'data'),
    Input('save-button', 'n_clicks'),
    State('pm10-plot', 'figure')
)
def save_as_jpeg(n_clicks, fig):
    if n_clicks > 0:
        return dcc.send_bytes(lambda x: go.Figure(fig).write_image(x, format="jpeg", scale=2), "visualization.jpg")

if __name__ == '__main__':
    app.run_server(debug=True)
```
